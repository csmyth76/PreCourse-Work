```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

df = pd.read_csv('HR_comma_sep.csv')
dept_counts = Counter(df['sales'])
lists = sorted(dept_counts.items()) # sorted by key, return a list of tuples
x, y = zip(*lists) # unpack a list of pairs into two tuples
plt.bar(x, y)
plt.xticks(rotation=90) # fix labels (unreadable without this)
plt.show()
```

What department has the highest percentage of satisfied employees?
    technical

```python

df_only_satisfied = df.loc[lambda x: df['satisfaction_level'] == 1, :]
dept_satisfied_counts = Counter(df_only_satisfied['sales'])

d1=dept_counts 
d2=dept_satisfied_counts
dept_satisfied_percent = {k: (d2[k]/d1[k])*100 for k in d1.keys() & d2}

lists = sorted(dept_satisfied_percent.items()) # sorted by key, return a list of tuples
x, y = zip(*lists) # unpack a list of pairs into two tuples
plt.bar(x, y)
plt.xticks(rotation=90) # fix labels (unreadable without this)
plt.show()
```
