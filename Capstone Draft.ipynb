```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

df = pd.read_csv('HR_comma_sep.csv')
dept_counts = Counter(df['sales'])
lists = sorted(dept_counts.items()) # sorted by key, return a list of tuples
x, y = zip(*lists) # unpack a list of pairs into two tuples
plt.bar(x, y)
plt.xticks(rotation=90) # fix labels (unreadable without this)
plt.show()
```
