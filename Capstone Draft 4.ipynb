Initial Question: What factors contribute to workplace satisfaction?
- there is a correlation between too many hours and lowest satisfactory rating
- there is a correlation between high employee evaluations and lowest satisfactory

Data:
    Last evaluation
    Number of projects
    Average monthly hours
    Time spent at the company
    Whether they have had a work accident
    Whether they have had a promotion in the last 5 years
    Departments (column sales)
    Salary
    Whether the employee has left

Supporting Question: 
Does monthly hours contribute to workplace satisfaction? Yes

- Most people who gave the lowest satisfaction rating worked approximately 240 to over 300

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import operator
from collections import Counter

df = pd.read_csv('HR_comma_sep.csv')

plt.scatter(x=df['satisfaction_level'], y=df['average_montly_hours'], s=1)
plt.show()
```

Supporting Question: 
Does the employee's last evaluation rating contribute to workplace satisfaction? ? Yes

- Most people who gave the lowest satisfaction scored approximately .8 to 1 in the last employee
evaluation

```python
plt.scatter(x=df['satisfaction_level'], y=df['last_evaluation'], s=1)
plt.show()
```

Supporting Question: Does salary effect satisfaction level? No
    - Satisfaction levels are similar for each salary range

```python
df.boxplot(
    column='satisfaction_level', 
    by='salary',
    showmeans=True
)
plt.show()
```

Supporting Question: Does having a work accident effect satisfaction level? No
    - Satisfaction levels are similar regardless of whether an employee had an accident or not

```python
df.boxplot(
    column='satisfaction_level', 
    by='Work_accident',
    showmeans=True
)
plt.show()
```

Supporting Question: Does department effect satisfaction level? No
    - Satisfaction levels are similar regardless of which department they worked in

```python
df.boxplot(
    column='satisfaction_level', 
    by='sales',
    showmeans=True
)
plt.xticks(rotation=90)
plt.show()
```

Supporting Question: Does having a promotion in the past 5 years effect satisfaction level? No
    - Satisfaction levels are slightly higher after having a promotion but still similar 

```python
df.boxplot(
    column='satisfaction_level', 
    by='promotion_last_5years',
    showmeans=True
)
plt.show()
```

What factors are related to people leaving?

Where salaries higher for those that stayed vs those that left?

```python
df_left = df.loc[lambda x: df['left'] == 1, :]
df_no_left = df.loc[lambda x: df['left'] == 0, :]

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
df_left['promotion_last_5years'].value_counts().plot(kind='bar')
plt.subplot(1, 2, 2)
df_no_left['promotion_last_5years'].value_counts().plot(kind='bar')
plt.show()
```

Where salaries higher for those that stayed vs those that left?  Significantly more people who made low and med salaries left vs high salaries

```python
df_left = df.loc[lambda x: df['left'] == 1, :]
df_no_left = df.loc[lambda x: df['left'] == 0, :]

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
df_left['salary'].value_counts().plot(kind='bar')
plt.subplot(1, 2, 2)
df_no_left['salary'].value_counts().plot(kind='bar')
plt.show()
```

Where the number of projects correlated with people leaving? Yes, more people who only worked on 2 projects left

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
df_left['number_project'].value_counts().plot(kind='bar')
plt.subplot(1, 2, 2)
df_no_left['number_project'].value_counts().plot(kind='bar')
plt.show()
```

Are work accidents a factor in people leaving?  No

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
df_left['Work_accident'].value_counts().plot(kind='bar')
plt.subplot(1, 2, 2)
df_no_left['Work_accident'].value_counts().plot(kind='bar')
plt.show()
```
